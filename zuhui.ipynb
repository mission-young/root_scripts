```c++
void pege(int id,int g1,int g2,int cond=1,int decaymark=0){
    c.Clear();
    TH2F *h1;
    TH2F *h2;
    TH2F *h3;
    TH2F *h4;
    if(id==1&&cond==1){
        // le1lea1
        tree->Draw("le1:xiagea>>h1(4000,0,4000,1000,100,10100)","cond==1&&bz==1&&dtime<91*5");
        tree->Draw("le1:xiagea>>h2(4000,0,4000,1000,100,10100)","cond==1&&bz==1&&dtime>5000-91*5");
        h1=(TH2F*)gROOT->FindObject("h1");
        h2=(TH2F*)gROOT->FindObject("h2");
        h1->Add(h1,h2,1,-1);
        
        // CUTG
        tree->Draw("le1:xiagea>>h3(4000,0,4000,1000,100,10100)","cond==2&&decaymark==8&&CUTG&&bz==1&&dtime<91*5");
        tree->Draw("le1:xiagea>>h4(4000,0,4000,1000,100,10100)","cond==2&&decaymark==8&&CUTG&&bz==1&&dtime>5000-91*5");
        h3=(TH2F*)gROOT->FindObject("h3");
        h4=(TH2F*)gROOT->FindObject("h4");
        h1->Add(h1,h3,1,1);
        h1->Add(h1,h4,1,-1);
        h1->SetName("hh11");

    }
    if(id==2&&cond==1){
        //Si beabe
        tree->Draw("bea:xiagea>>h1(4000,0,4000,1000,100,10100)","cond==1&&bz==2&&dtime<91*5&&ve[0]<200");
        tree->Draw("bea:xiagea>>h2(4000,0,4000,1000,100,10100)","cond==1&&bz==2&&dtime>5000-91*5&&ve[0]<200");
        h1=(TH2F*)gROOT->FindObject("h1");
        h2=(TH2F*)gROOT->FindObject("h2");
        h1->Add(h1,h2,1,-1);
        h1->SetName("hh21");
    }
    if(id==1&&cond==2&&decaymark==7){
        //decaymark7Sicond=2id=1Si Si
        tree->Draw("be:xiagea>>h1(4000,0,4000,1000,100,10100)","cond==2&&bz==1&&decaymark==7&&dtime<91*5");
        tree->Draw("be:xiagea>>h2(4000,0,4000,1000,100,10100)","cond==2&&bz==1&&decaymark==7&&dtime>5000-91*5");
        h1=(TH2F*)gROOT->FindObject("h1");
        h2=(TH2F*)gROOT->FindObject("h2");
        h1->Add(h1,h2,1,-1);
        h1->SetName("hh127");
    }
        if(id==1&&cond==2&&decaymark==8){
        // be(le1le2)  le1 & lea2.  Si
        // CUTG1300keV2500keV
        tree->Draw("0.987434*(le1+lea2)+97.7:xiagea>>h1(4000,0,4000,1000,100,10100)","cond==2&&bz==1&&decaymark==8&&!CUTG&&dtime<91*5");
        tree->Draw("0.987434*(le1+lea2)+97.7:xiagea>>h2(4000,0,4000,1000,100,10100)","cond==2&&bz==1&&decaymark==8&&!CUTG&&dtime>5000-91*5");
        h1=(TH2F*)gROOT->FindObject("h1");
        h2=(TH2F*)gROOT->FindObject("h2");
        h1->Add(h1,h2,1,-1);
        h1->SetName("hh128");
    }
        if(id==2&&cond==2&&decaymark==8){
        tree->Draw("0.987434*(le1+lea2)+97.7:xiagea>>h1(4000,0,4000,1000,100,10100)","cond==2&&bz==2&&decaymark==8&&ve[0]<200&&dtime<91*5");
        tree->Draw("0.987434*(le1+lea2)+97.7:xiagea>>h2(4000,0,4000,1000,100,10100)","cond==2&&bz==2&&decaymark==8&&ve[0]<200&&dtime>5000-91*5");
        h1=(TH2F*)gROOT->FindObject("h1");
        h2=(TH2F*)gROOT->FindObject("h2");
        h1->Add(h1,h2,1,-1);
        h1->SetName("hh228");
    }
    int bin1,bin2,de;
    de=g1-g2;
    bin1=h1->GetXaxis()->FindBin(g1);
    bin2=h1->GetXaxis()->FindBin(g2);
    TH1D *p1=h1->ProjectionY("p1",bin1,bin2);
    bin1=h1->GetXaxis()->FindBin(g1-10*de-1);
    bin2=h1->GetXaxis()->FindBin(g1-1);
    TH1D *p2=h1->ProjectionY("p2",bin1,bin2);
    bin1=h1->GetXaxis()->FindBin(g2+1);
    bin2=h1->GetXaxis()->FindBin(g2+de*10+1);
    TH1D *p3=h1->ProjectionY("p3",bin1,bin2);
    p1->Add(p1,p2,1,-0.05);
    p1->Add(p1,p3,1,-0.05);
    TString title;
    title.Form("pe%d with gcut[ %d-%d keV ] and all background sub cond=%d, decaymark=%d",id,g1,g2,cond,decaymark);
    p1->SetTitle(title.Data());
    p1->Draw();
    gPad->SetLogy();
    c.Draw();
}
```
