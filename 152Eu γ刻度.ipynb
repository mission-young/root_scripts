```c++
%jsroot on
TCanvas c;
TFile f("eu152eff.root");
tree->Draw("gea>>g(10000,0,2000)");
c.Draw();
```

```c++
Double_t f1(Double_t *x,Double_t *p){
    return p[0]*TMath::Gaus(x[0],p[1],p[2])+p[3]*TMath::Erf((p[1]-x[0])/(sqrt(2)*p[2]))+p[4]+p[5]*x[0];
}
```

```c++
TF1 *f=new TF1("f",f1,0,1500,6);
f->SetParameters(1e4,121,2,4000,2.6e4,-80);
auto s=g->Fit(f,"S","",107,140);
c.Draw();
```

```c++
Double_t f2(Double_t *x,Double_t *p){
    return p[3]*TMath::Erf((p[1]-x[0])/(sqrt(2)*p[2]))+p[4]+p[5]*x[0];
}
TF1 *bkg=new TF1("bkg",f2,0,1500,6);
```

```c++
void res(int l,int r){
    bkg->SetParameters(s->GetParams());
    cout<<" "<<g->Integral(l*5,r*5)<<'\n'<<" "<<bkg->Integral(l,r)*5<<endl;
}
res(110,130);
```
