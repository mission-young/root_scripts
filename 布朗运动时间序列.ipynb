```c++
%jsroot on
TCanvas c;
TMultiGraph *mg=new TMultiGraph();
double b[1000];
double t[1000];
```

```c++
TRandom3 R;
R.SetSeed(0);
cout<<R.Gaus()<<endl;
```

# 布朗运动序列

```c++
memset(b,0,sizeof(b));
memset(t,0,sizeof(t));
for(int i=0;i<1000;i++){
    t[i]=i+1;
    for(int j=i;j>=0;j--){
        b[i]+=R.Gaus();
    }
}
TGraph *Bt=new TGraph(1000,t,b);
Bt->Draw("A");
c.Draw();
```

# 极大值序列

```c++
int e[999];
double flag=0;
for(int i=0;i<999;i++){
    if(abs(b[i+1]-b[i])<1e-5) flag=0;
    if(b[i+1]-b[i]>0) flag=1;
    if(b[i+1]-b[i]<0) flag=-1;
    e[i]=flag;
}
vector<int> maxp;
vector<int> minp;

if(e[0]<0) maxp.push_back(0);
if(e[0]>0) minp.push_back(0);
for(int i=0;i<998;i++){
     if(!(e[i]==0&&e[i+1]==0)&&e[i]>=0&&e[i+1]<=0){
         maxp.push_back(i+1);
     }else if(!(e[i]==0&&e[i+1]==0)&&e[i]<=0&&e[i+1]>=0){
        minp.push_back(i+1);
     }else{
         continue;
     }
}
if(e[998]>0) maxp.push_back(999);
if(e[998]<0) minp.push_back(999);

TGraph *gmax=new TGraph();
TGraph *gmin=new TGraph();
for(int i=0;i<maxp.size();i++){
    gmax->SetPoint(i,maxp[i]+1,b[maxp[i]]);
}
for(int i=0;i<minp.size();i++){
    gmin->SetPoint(i,minp[i]+1,b[minp[i]]);
}
c.Clear();
gmax->Draw("A*");
c.Draw();
```

# 极小值序列

```c++
c.Clear();
gmin->Draw("A*");
c.Draw();
```

# 极大极小值序列检验

```c++
Bt->SetMarkerStyle(21);
Bt->SetMarkerColor(1);
Bt->SetFillStyle(0);
mg->Add(Bt,"PL");
gmin->SetMarkerStyle(22);
gmin->SetMarkerColor(2);
gmin->SetFillStyle(0);
mg->Add(gmin,"P");
gmax->SetMarkerStyle(23);
gmax->SetMarkerColor(3);
gmax->SetFillStyle(0);
mg->Add(gmax,"P");
c.Clear();
mg->Draw();
Bt->SetName("Brownian Motion");
Bt->SetTitle("Brownian Motion");
gmin->SetName("local min");
gmax->SetName("local max");
gmin->SetTitle("local min");
gmax->SetTitle("local max");
c.BuildLegend();
mg->GetYaxis()->SetRangeUser(-200,200);
mg->GetXaxis()->SetLimits(0,1001);
c.Draw();
```

```c++
TMultiGraph *localp=new TMultiGraph();
localp->Add(gmin,"P");
localp->Add(gmax,"P");
c.Clear();
localp->Draw();
c.Draw();
```

```c++
int s=3;
int *local=new int[gmin->GetN()+gmax->GetN()];
int bias;
if(gmin->GetPointX(0)>gmax->GetPointY(0)){
    bias=1;
} else{
    bias=0;
}
for(int i=0;i<gmin->GetN();i++){
    local[2*i+bias]=gmin->GetPointX(i);
}
for(int i=0;i<gmax->GetN();i++){
    local[2*i+!(bias)]=gmax->GetPointX(i);
}
```

```c++
TGraph *test=new TGraph;
for(int i=0;i<gmin->GetN()+gmax->GetN();i++){
    test->SetPoint(i,i+1,local[i]);
}
c.Clear();
test->Draw("A*");
c.Draw();
```
